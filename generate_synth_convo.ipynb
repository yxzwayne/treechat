```python
import json
import random
import uuid
from datetime import datetime
```

a message should have:
- id 
- created_time
- role
- content_type
- content
- media
- parent_id
- children_id (this is a list)

```python
import string


def generate_random_message(length):
    return "".join(random.choices(string.ascii_letters + string.digits, k=length))


data = []

for _ in range(10):
    convo_id = str(uuid.uuid4())
    # create the chats within this convo
    # randomly determine how many turns will the user have, between 2-5
    n_usr_turn = random.randint(2, 5)
    for msg_i in range(n_usr_turn):
        if msg_i % 2 == 0:
            role = "user"
        else:
            role = "assistant"
        # create messages:
        msg_id = str(uuid.uuid4())
        msg = {
            "id": msg_id,
            "created_time": datetime.now().isoformat(),
            "role": role,
            "content_type": "text",
            "content": generate_random_message(random.randint(22, 84)),
            "media": [],
            "parent_id": None,
            "children_id": [],
        }

    convo = {
        "id": convo_id,
        "messages": [msg],
    }
    data.append(convo)


print(data)
```
